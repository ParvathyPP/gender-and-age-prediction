```python
age_group = []
def func_age_group(age):
    age = int(age)
    if age >= 0 and age < 18:
        op = "(0 - 17)"
        return op
    elif age <= 30:
        op = "(18 - 29)"
        return op
    elif age <= 50:
        op = "(30 - 50)"
        return op
    elif age <= 75:
        op = "(51 - 75)"
        return op
    else:
        op = "(>76)"
        return op
for i in ages:
    group = func_age_group(i)
    age_group.append(group)
```

```python
df = pd.DataFrame(ages, columns = ['ages'])
df['genders'] = genders
df['age_group'] = age_group
```
