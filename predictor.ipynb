```python
#import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import tensorflow as tf
import cv2
import os
from tensorflow import keras
```

```python
path = "C:/Users/USER/Documents/kaggle_projects/Gender and Age Prediction/gender-and-age-prediction/datasets/UTKFace"
files = os.listdir(path)
```

```python
images = []
ages = []
genders = []
for img in files:
  total=path+'/'+img
  image = cv2.imread(total, 0)
  image = cv2.resize(image, dsize = (48, 48))
  image = image.reshape((image.shape[0], image.shape[1], 1))
  images.append(image)
  ages.append(img.split('_')[0])
  genders.append(img.split('_')[1])
```

```python
img = images[30]
cv2.imshow("test", img)
cv2.waitKey()
cv2.destroyAllWindows()
print(ages[30])
print(genders[30])
```

```python
values ,counts = np.unique(genders, return_counts= True)
print(values)
print(counts)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
values = ['male', 'female']
ax.bar(values, counts)
```
